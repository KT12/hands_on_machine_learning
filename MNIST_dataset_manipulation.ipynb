```python
mnist = input_data.read_data_sets("/tmp/data/")
```

```python
X_train, y_train = mnist.train.images, mnist.train.labels
```

```python
enumerated_y = list(enumerate(y_train))
```

```python
indices = [index for index, label in enumerated_y if label < 5]
```
